```python
import uproot
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Signal (73 columns)

sig = uproot.open('/Users/puntong/Desktop/MLinHEP/data/yhhbb_s.root')
sig.classnames()
```

```python
sig['mssmhbb'].keys()
```

```python
tree_sig = sig['mssmhbb']
arr_sig = np.array(tree_sig).T
df_sig = pd.DataFrame(arr_sig, columns=tree_sig.keys())

df_sig.insert(0, 'signal', 1.0)

df_sig   # Signal (73 columns)
```

```python
df_sig['flavorj1'].value_counts()
```

```python
# Background file 1 (38 columns)

bg1 = uproot.open('/Users/puntong/Desktop/MLinHEP/data/QCD_bEnriched_HT100toInf.root')
bg1.classnames()
```

```python
tree_bg1 = bg1['mssmhbb']
arr_bg1 = np.array(tree_bg1).T
df_bg1 = pd.DataFrame(arr_bg1, columns=tree_bg1.keys())

df_bg1.insert(0, 'signal', 0.0)

df_bg1  # Background file 1 (38 columns)
```

```python
# Background file 2 (38 columns)

bg2 = uproot.open('/Users/puntong/Desktop/MLinHEP/data/SUSYGluGluToBBH_M-300.root')
bg2.classnames()
```

```python
tree_bg2 = bg2['mssmhbb']
arr_bg2 = np.array(tree_bg2).T
df_bg2 = pd.DataFrame(arr_bg2, columns=tree_bg2.keys())

df_bg2.insert(0, 'signal', 0.0)

df_bg2  # Background file 2 (38 columns)
```

```python
# Signal (38 columns same as Background) 

df_sig_new = df_sig[df_bg2.keys()]
df_sig_new 
```

```python
# combine Signal and Background (only 38 columns)

df_data = pd.concat([df_sig_new, df_bg1, df_bg2], axis=0, ignore_index=True)
data = df_data
data 
```

```python
# Signal vs Background Density plot (38 columns)

fig = plt.figure(figsize=(40,20))
m=0
for i in data:
    if i != 'signal':
        fig.add_subplot(6, 7, m+1)
        plt.hist(data[i][:300000], bins=50, density=True, histtype='step', label='signal')
        plt.hist(data[i][300000:779728], bins=50, density=True, histtype='step', label='background')
        plt.xlabel(i)
        plt.ylabel('Density')
        plt.legend()
        m+=1    
plt.tight_layout()
#plt.legend()
plt.show()
```

```python
# Signal vs Background Frequency plot (38 columns)

fig = plt.figure(figsize=(40,20))
m=0
for i in data:
    if i != 'signal':
        fig.add_subplot(6, 7, m+1)
        plt.hist(data[i][:300000], bins=50, density=False, histtype='step', label='signal')
        plt.hist(data[i][300000:779728], bins=50, density=False, histtype='step', label='background')
        plt.xlabel(i)
        plt.ylabel('Frequency')
        plt.legend()
        m+=1    
plt.tight_layout()
#plt.legend()
plt.show()
```

```python
# Signal Density plot (others)

other = list(set(list(df_sig.keys())) - set(list(df_bg2.keys())))
df_other = df_sig[other]

fig = plt.figure(figsize=(40,20))
m=0
for i in df_other:
    if i != 'signal':
        fig.add_subplot(6, 7, m+1)
        plt.hist(df_other[i], bins=50, density=True, histtype='step', label='signal')
        plt.xlabel(i)
        plt.ylabel('Density')
        plt.legend()
        m+=1    
plt.tight_layout()
#plt.legend()
plt.show()
```

```python
# Signal Frequency plot (others)

other = list(set(list(df_sig.keys())) - set(list(df_bg2.keys())))
df_other = df_sig[other]

fig = plt.figure(figsize=(40,20))
m=0
for i in df_other:
    if i != 'signal':
        fig.add_subplot(6, 7, m+1)
        plt.hist(df_other[i], bins=50, density=False, histtype='step', label='signal')
        plt.xlabel(i)
        plt.ylabel('Frequency')
        plt.legend()
        m+=1    
plt.tight_layout()
#plt.legend()
plt.show()
```
